```python
# ðŸ“¦ Step 1: Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import files
import io # Import the io module

# ðŸ“¥ Step 2: Upload CSV file
print("Upload your accident_data.csv file:")
uploaded = files.upload()

# ðŸ“„ Step 3: Read uploaded file and check if data was loaded
df = None # Initialize df
for filename, content in uploaded.items(): # Iterate through uploaded items (filename and content)
    # Read the uploaded content directly into a DataFrame using io.StringIO
    df = pd.read_csv(io.StringIO(content.decode('utf-8')))
    print(f"File '{filename}' uploaded and read.")
    break # Assuming only one file is uploaded

# Check if the DataFrame was successfully loaded and is not empty
if df is None or df.empty:
    print("Error: No data was loaded from the uploaded file.")
else:
    # ðŸ” Step 4: Prediction Function
    def predict_accident(row):
        risk_score = 0

        # Check if columns exist before accessing them
        if 'Weather' in row and row['Weather'] == 'Rainy':
            risk_score += 2
        if 'Road_Condition' in row and row['Road_Condition'] == 'Wet':
            risk_score += 2
        if 'Time_of_Day' in row and row['Time_of_Day'] == 'Night':
            risk_score += 1
        # Add checks for numeric columns and non-null values
        if 'Traffic_Density' in row and pd.notnull(row['Traffic_Density']) and row['Traffic_Density'] > 2:
            risk_score += 1
        if 'Driver_Alcohol' in row and pd.notnull(row['Driver_Alcohol']) and row['Driver_Alcohol'] == 1:
            risk_score += 3
        if 'Speed_Limit' in row and pd.notnull(row['Speed_Limit']) and row['Speed_Limit'] > 80:
            risk_score += 1
        if 'Number_of_Vehicles' in row and pd.notnull(row['Number_of_Vehicles']) and row['Number_of_Vehicles'] > 2:
            risk_score += 1

        return 'High Risk' if risk_score >= 5 else 'Low Risk'

    # ðŸ’¡ Step 5: Predict for each row
    df['Prediction'] = df.apply(predict_accident, axis=1)

    # ðŸ§¾ Step 6: Display result
    print("Predictions completed:")
    # Ensure the columns to display exist in the DataFrame
    display_cols = [col for col in ['Weather', 'Road_Condition', 'Time_of_Day', 'Prediction'] if col in df.columns]
    display(df[display_cols])

    # ðŸ’¾ Step 7: Save output CSV
    df.to_csv('accident_predictions.csv', index=False)
    files.download('accident_predictions.csv')

    # ðŸ“Š Step 8: Visualizations
    # Wrap plotting code in a try-except block or add checks to prevent errors if data is not suitable
    if 'Prediction' in df.columns:
        ## 1. Count of Predictions
        plt.figure(figsize=(6,4))
        sns.countplot(data=df, x='Prediction', palette='coolwarm')
        plt.title("Count of Risk Predictions")
        plt.xlabel("Prediction")
        plt.ylabel("Count")
        plt.grid(axis='y')
        plt.show()

        if 'Weather' in df.columns:
            ## 2. Prediction by Weather
            plt.figure(figsize=(8,5))
            sns.countplot(data=df, x='Weather', hue='Prediction', palette='viridis')
            plt.title("Prediction by Weather Condition")
            plt.ylabel("Number of Records")
            plt.grid(axis='y')
            plt.show()

        if 'Traffic_Density' in df.columns:
             ## 3. Risk vs Traffic Density
            plt.figure(figsize=(8,5))
            sns.boxplot(data=df, x='Prediction', y='Traffic_Density', palette='Set2')
            plt.title("Traffic Density vs Risk")
            plt.grid(True)
            plt.show()

        if 'Speed_Limit' in df.columns:
            ## 4. Speed Distribution by Risk
            plt.figure(figsize=(8,5))
            sns.histplot(data=df, x='Speed_Limit', hue='Prediction', kde=True, element='step', bins=10)
            plt.title("Speed Distribution by Prediction")
            plt.grid(True)
            plt.show()
```
