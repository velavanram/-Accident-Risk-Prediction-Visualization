```python
# 📦 Step 1: Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import files

# 📥 Step 2: Upload CSV file
print("Upload your accident_data.csv file:")
uploaded = files.upload()

# 📄 Step 3: Read uploaded file
for filename in uploaded:
    df = pd.read_csv(filename)

# 🔍 Step 4: Prediction Function
def predict_accident(row):
    risk_score = 0

    if row['Weather'] == 'Rainy':
        risk_score += 2
    if row['Road_Condition'] == 'Wet':
        risk_score += 2
    if row['Time_of_Day'] == 'Night':
        risk_score += 1
    if row['Traffic_Density'] > 2:
        risk_score += 1
    if row['Driver_Alcohol'] == 1:
        risk_score += 3
    if row['Speed_Limit'] > 80:
        risk_score += 1
    if row['Number_of_Vehicles'] > 2:
        risk_score += 1

    return 'High Risk' if risk_score >= 5 else 'Low Risk'

# 💡 Step 5: Predict for each row
df['Prediction'] = df.apply(predict_accident, axis=1)

# 🧾 Step 6: Display result
print("Predictions completed:")
display(df[['Weather', 'Road_Condition', 'Time_of_Day', 'Prediction']])

# 💾 Step 7: Save output CSV
df.to_csv('accident_predictions.csv', index=False)
files.download('accident_predictions.csv')

# 📊 Step 8: Visualizations

## 1. Count of Predictions
plt.figure(figsize=(6,4))
sns.countplot(data=df, x='Prediction', palette='coolwarm')
plt.title("Count of Risk Predictions")
plt.xlabel("Prediction")
plt.ylabel("Count")
plt.grid(axis='y')
plt.show()

## 2. Prediction by Weather
plt.figure(figsize=(8,5))
sns.countplot(data=df, x='Weather', hue='Prediction', palette='viridis')
plt.title("Prediction by Weather Condition")
plt.ylabel("Number of Records")
plt.grid(axis='y')
plt.show()

## 3. Risk vs Traffic Density
plt.figure(figsize=(8,5))
sns.boxplot(data=df, x='Prediction', y='Traffic_Density', palette='Set2')
plt.title("Traffic Density vs Risk")
plt.grid(True)
plt.show()

## 4. Speed Distribution by Risk
plt.figure(figsize=(8,5))
sns.histplot(data=df, x='Speed_Limit', hue='Prediction', kde=True, element='step', bins=10)
plt.title("Speed Distribution by Prediction")
plt.grid(True)
plt.show()
```
